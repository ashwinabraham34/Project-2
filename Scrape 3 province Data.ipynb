```python
from splinter import Browser

executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path)
```

```python
url = 'https://www.alberta.ca/stats/covid-19-alberta-statistics.htm'
browser.visit(url)
```

```python
browser.click_link_by_partial_text('Data export')
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
from splinter import Browser
```

```python
table = soup.find('div', class_='dataTables_scrollBody')
table_rows = table.find_all('tr')
```

```python
# Function to Click Next Page        
def click_next_page():
    browser.click_link_by_partial_text('Next')
res = []
page = 1
while page <146:
    print ('Page: %s' %page)
    table = soup.find('div', class_='dataTables_scrollBody')
    table_rows = table.find_all('tr')

    for x in table_rows:
        td = x.find_all('td')
        row = [x.text.strip() for x in td if x.text.strip()]
        if row:
            res.append(row)

    ## Your code to Parse the Tables ## 

    try:
        click_next_page()
        page += 1
    except:
        print ('You are at the end')
        ##num_of_pages == False


    ##time.sleep(0)
```

```python
df = pd.DataFrame(res, columns=["index1","Date Reported","Alberta Health Zone","Gender","Age Group","Case Status", "Case Type"])
df
```
